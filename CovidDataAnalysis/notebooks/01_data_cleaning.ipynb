```python
df = df[~df["location"].isin(["World","Europe","North America","South America","Asia","Africa"])]
df['date'] = pd.to_datetime(df['date'])
```

```python
country_nan_percentage(df, "location", target_col="new_deaths")
```

```python
columns_nan_percentage(df)
```

```python
keep_country = []
for col in df["location"].unique():
    size = df.loc[df["location"] == col, "new_deaths"].shape[0]
    nan_size = df.loc[df["location"] == col, "new_deaths"].isna().sum()
    if (nan_size / size) * 100 < 20:
        keep_country.append(col)

print(" - ".join(keep_country))
```

```python
keep_columns = [
    'continent', 'location', 'date',
    'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'population', 'population_density'
]
```

```python
df_filtered = df.loc[df["location"].isin(keep_country) , keep_columns]
df_filtered.sort_values(by=['location', 'date'], ascending=[True, True], inplace=True)
df_filtered
```

```python
df_filtered.info()
```

```python
df_filtered.columns = ['continent', 'country', 'date',
                       'total_cases', 'new_cases','total_deaths',
                       'new_deaths', 'population', 'population_density']
```
